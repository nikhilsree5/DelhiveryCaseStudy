##Feature Creation


```python
df1.head()
```

```python
df1['od_start_time']=pd.to_datetime(df1['od_start_time'])
df1['od_end_time']=pd.to_datetime(df1['od_end_time'])
```

```python
df1['od_time_diff_hour']=(df1['od_end_time']-df1['od_start_time']).dt.total_seconds() / 3600
```

```python
df1
```

```python
df1['source_name']=df1['source_name'].fillna('NA_NA_NA(NA)')

df1['destination_name']=df1['destination_name'].fillna('NA_NA_NA(NA)')
```

```python
df1['source_state']=df1['source_name'].apply(lambda X: X.split('(')[1].split(')')[0])
df1['source_city']=df1['source_name'].apply(lambda X: X.split('_')[0])
df1['destination_state']=df1['destination_name'].apply(lambda X: X.split('(')[1].split(')')[0])
df1['destination_city']=df1['destination_name'].apply(lambda X: X.split('_')[0])
```

```python
df1
```
